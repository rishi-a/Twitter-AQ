## Monthly Linear Fit

```python
from numpy.polynomial.polynomial import polyfit
import matplotlib.pyplot as plt
```

```python
pmLevels = df3.drop(['Unnamed: 0', 'time'], axis=1)
pmLevels1 = pmLevels['2019-01-01':'2019-12-31']
pmLevels1 = pmLevels1.resample('1D').mean()
pmLevels1 = pmLevels1.dropna()
```

```python
df = pd.read_csv('all-delhi-tweets-2019.csv')
df['Freq'] = 1
df.index = pd.to_datetime(df['date'], errors='coerce')
tweets = df[['Freq']]
tweets = tweets.resample('1D').sum()
#remove spurious entries
tweets.drop(tweets.tail(1).index,inplace=True)
```

```python
merge=pd.merge(tweets,pmLevels1, how='inner', left_index=True, right_index=True)
```

```python
merge
```

```python
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
monthIndex = 0
for i in range(1,13):
    month_mask = merge.index.map(lambda x: x.month) == i
    df_month = pd.DataFrame()
    df_month = merge[month_mask]
    print(df_month.shape)
    #print(df_month)
    
    
    # Split the data into training/testing sets
    df_month_X_train = np.array(df_month['Freq'][0:]).reshape(-1,1)
    df_month_X_test = np.array(df_month['Freq'][0:]).reshape(-1,1)

    # Split the targets into training/testing sets
    df_month_y_train = np.array(df_month['PM2.5'][0:]).reshape(-1,1)
    df_month_y_test = np.array(df_month['PM2.5'][0:]).reshape(-1,1)
    
    # Create linear regression object
    regr = linear_model.LinearRegression()
    try:
        # Train the model using the training sets
        regr.fit(df_month_X_train, df_month_y_train)

        # Make predictions using the testing set
        df_month_y_pred = regr.predict(df_month_X_test)
    except:
        continue
    
    # The coefficients
    print('Coefficients: \n', regr.coef_)
    # The mean squared error
    print('Mean squared error: %.2f'% mean_squared_error(df_month_y_test, df_month_y_pred))
    
    # Plot outputs
    plt.scatter(df_month_X_test, df_month_y_test,  color='black')
    plt.plot(df_month_X_test, df_month_y_pred, color='blue', linewidth=3)
    
    month = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
    
    plt.title("%s"%month[monthIndex])
    plt.ylabel("PM$_{2.5}$ Values")
    plt.xlabel("# Tweets")

    monthIndex+=1

    plt.show()
```

```python
#Original Author: Prof. Nipun Batra
# nipunbatra.github.io
from math import sqrt 
import matplotlib
SPINE_COLOR = 'gray'
def latexify(fig_width=None, fig_height=None, columns=1):
    """Set up matplotlib's RC params for LaTeX plotting.
    Call this before plotting a figure.

    Parameters
    ----------
    fig_width : float, optional, inches
    fig_height : float,  optional, inches
    columns : {1, 2}
    """

    # code adapted from http://www.scipy.org/Cookbook/Matplotlib/LaTeX_Examples

    # Width and max height in inches for IEEE journals taken from
    # computer.org/cms/Computer.org/Journal%20templates/transactions_art_guide.pdf

    assert(columns in [1,2])

    if fig_width is None:
        fig_width = 3.39 if columns==1 else 6.9 # width in inches

    if fig_height is None:
        golden_mean = (sqrt(5)-1.0)/2.0    # Aesthetic ratio
        fig_height = fig_width*golden_mean # height in inches

    MAX_HEIGHT_INCHES = 8.0
    if fig_height > MAX_HEIGHT_INCHES:
        print("WARNING: fig_height too large:" + fig_height + 
              "so will reduce to" + MAX_HEIGHT_INCHES + "inches.")
        fig_height = MAX_HEIGHT_INCHES

    params = {'backend': 'ps',
              'text.latex.preamble': ['\\usepackage{gensymb}'],
              'axes.labelsize': 8, # fontsize for x and y labels (was 10)
              'axes.titlesize': 8,
              'font.size': 8, # was 10
              'legend.fontsize': 8, # was 10
              'xtick.labelsize': 8,
              'ytick.labelsize': 8,
              'text.usetex': True,
              'figure.figsize': [fig_width,fig_height],
              'font.family': 'serif'
    }

    matplotlib.rcParams.update(params)


def format_axes(ax):

    for spine in ['top', 'right']:
        ax.spines[spine].set_visible(False)

    for spine in ['left', 'bottom']:
        ax.spines[spine].set_color(SPINE_COLOR)
        ax.spines[spine].set_linewidth(0.5)

    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')

    for axis in [ax.xaxis, ax.yaxis]:
        axis.set_tick_params(direction='out', color=SPINE_COLOR)

    return ax
```

```python
from numpy.random import rand
import matplotlib
import matplotlib.pyplot as plt

# create all axes we need
ax0 = plt.subplot(211)
ax1 = ax0.twinx()
ax2 = plt.subplot(221)
ax3 = ax2.twinx()

# share the secondary axes
ax1.get_shared_y_axes().join(ax1, ax3)

ax0.plot(rand(1) * rand(10),'r')
ax1.plot(10*rand(1) * rand(10),'b')
ax2.plot(3*rand(1) * rand(10),'g')
ax3.plot(10*rand(1) * rand(10),'y')
```
