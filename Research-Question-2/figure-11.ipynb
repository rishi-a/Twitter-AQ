**Generate Figure 9 of the paper**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import latexify as lt
```

```python
df3 = pd.read_csv('pm25/pm25-data-BE.csv')
df3.index = pd.to_datetime(df3['time'], format="%Y-%m-%d:%H", errors='coerce')
df3 = df3.drop(['Unnamed: 0', 'time'], axis=1)
pmLevels = df3
pmLevels = pmLevels.resample('1D').mean()
pmLevels = pmLevels['2016-01-01':'2019-12-31']
```

```python
import matplotlib.colors as cl
#latexify(fig_width=3.7)
ax = pmLevels.plot()
ax.set_xlabel("Time")
ax.set_ylabel("PM$_{2.5}$ in $\mu g/m^3$")

plt.fill_between(pmLevels.index, 0,29, facecolor='#00cc00', alpha=0.9, label='Good')
plt.fill_between(pmLevels.index, 30,59, facecolor='#66cc00', alpha=0.9, label='Satisfactory')
plt.fill_between(pmLevels.index, 60,89, facecolor='#ffff00', alpha=0.9, label='Moderately polluted')
plt.fill_between(pmLevels.index, 90,119, facecolor='#ff9900', alpha=0.9, label='Poor')
plt.fill_between(pmLevels.index, 120,249, facecolor='#ff0000', alpha=0.9, label='Very Poor')
#plt.fill_between(pmLevels.index, 250,800, facecolor='#a52a2a', alpha=0.9, label='Beyond Very Poor')


ax.axhline(25, color="black", linewidth=2, linestyle='-.', label='WHO Limit')
ax.axhline(60, color="r", linewidth=2, linestyle='-.', label='India Limit')

plt.legend()
ax.legend(bbox_to_anchor=(0.5, 1.2), loc='upper center', ncol=2)


lt.format_axes(ax)
#plt.savefig('pm-vs-air-plots/yearly-pollution.pdf', bbox_inches='tight')
```
